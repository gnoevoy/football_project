```python
import pandas as pd
import numpy as np
import ast
import json
```

```python
file_path = r"/home/bars1k/football_project/data_json.json"
with open(file_path, 'r') as file:
    data = json.load(file)

df = pd.json_normalize(data["products"]) 
df.head(3)
```

```python
cols = [col for col in df.columns if "features." in col]
features = df[["product_id", *cols]]
features = features.melt(id_vars=["product_id"], value_vars=cols, var_name="feature", value_name="value").dropna()
features["feature"] = features["feature"].str.split(".").str[1]
features
```

```python
def get_features(df):
    cols = [col for col in df.columns if "features." in col]
    features_raw = df[["product_id", *cols]]
    features = pd.melt(features_raw, id_vars=["product_id"], value_vars=cols, var_name="feature", value_name="value").dropna()
    features["feature"] = features["feature"].str.split(".").str[-1]
    return features
```
